### Cleaning Rate Column

```python
data['rate'].unique()
```

```python
def cleanrate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
data['rate'] = data['rate'].apply(cleanrate)
data
```

### We can see Cost for two columns is not in integer data type, so let's clean it up and convert it to float datatype

```python
data['Cost for Two'].unique()
```

```python
def commaclean(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',' , '')
        return float(value)
    else:
        return float(value)

data['Cost for Two'] = data['Cost for Two'].apply(commaclean)
data['Cost for Two'].unique()    
```

```python
tempdf = data['rest_type'].value_counts(ascending = False)
tempdf
```

```python
data123 = data[(data['rest_type'].isin(tempdf[tempdf < 500].index)) & (data['rest_type'].str.count(',') > 0)]

# Save the resulting DataFrame to a new variable
data1234 = data123['rest_type'].value_counts()
data1234
```

```python
def handle_rest_type(value):
    if(value in data1234):
        return 'others'
    else:
        return value
data['rest_type'] = data['rest_type'].apply(handle_rest_type)
data['rest_type'].value_counts()
```
